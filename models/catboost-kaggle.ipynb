# Settings

```python
import numpy as np
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
```

```python
from datetime import date
import holidays
```

# Read in data

```python
train = pd.read_csv("../../train.csv")
```

```python
train_data = train[["is_delayed", "year", "month", "day", "sched_dep_time", "sched_arr_time", "carrier", "origin", "dest", "distance"]]
```

Make datime from year, month and day columns.
Make column with weekday names

```python
train_data['date'] = pd.to_datetime({'year': train_data['year'],
               'month': train_data['month'],
               'day': train_data['day']})
train_data['weekday'] = train_data['date'].dt.weekday_name
```

Find out if specific day was holiday

```python
us_holidays = holidays.UnitedStates()
```

```python
for date in train_data['date']:
     print(date in us_holidays)
```

```python
t = pd.DataFrame(list(map(lambda x: x in us_holidays, train_data['date'])))
```

```python
train_data= pd.concat([train_data, t], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data.drop(["is_delayed", "year", "date"], axis=1), train_data["is_delayed"], test_size=0.25)
```

```python
train_data
```

```python
X_train
```

```python
categorical_features = [4,5,6, 8]
```

Specify model parameters

```python
model = CatBoostClassifier(eval_metric="AUC",iterations=500, depth=10, learning_rate=0.15, 
                           loss_function='Logloss',l2_leaf_reg= 9, one_hot_max_size=31,
                           logging_level='Verbose')
```

Train model

```python
model.fit(X_train, y_train, cat_features=categorical_features, eval_set=(X_test, y_test))
```
